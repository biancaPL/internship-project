```python
import cv2
import matplotlib.pyplot as plt #we install two other packages that help us read the image
image = cv2.imread("/content/test.png") #we provide the path so that the system will get our image
plt.imshow(image)
plt.show()
```

```python
import pytesseract
extract = pytesseract.image_to_string(image) #we import the programm we install and we want to read the image, we perform OCR
print(extract)
```

```python
#open text file which is empty
text_file = open("/content/psa.txt", "w")

#write string to file
text_file.write(extract)

#close file
text_file.close()
```
